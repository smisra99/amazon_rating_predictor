# Data Cleaning

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
import re
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

def data_cleaning(row):

  row = row.lower()
  row = re.sub('[^A-Za-z]+', ' ', row)

  text_tokens = word_tokenize(row)
  # stop_words = stopwords.words('english')
  stop_words = set(stopwords.words('english'))

  list_clean_words = [word for word in text_tokens if not word in stop_words]

  lemmatize = WordNetLemmatizer()
  for i in range(len(list_clean_words)):
    list_clean_words[i] = lemmatize.lemmatize(list_clean_words[i])
  row = ' '.join(list_clean_words)
  return row

```

```python
df['FinalReviewText'] = df['FinalReviewText'].apply(lambda x: data_cleaning(x))
```

# Vectorize

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import sklearn.metrics as metrics
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
import sklearn.linear_model
X_train,X_test,Y_train, Y_test = train_test_split(df['FinalReviewText'], df['overall'], test_size=0.25, random_state=30)
print("Train: ",X_train.shape,Y_train.shape,"Test: ",(X_test.shape,Y_test.shape))
```

```python
vectorizer= TfidfVectorizer()
tf_x_train = vectorizer.fit_transform(X_train)
tf_x_test = vectorizer.transform(X_test)
```

# Different classifiers



```python
lr = sklearn.linear_model.LogisticRegression(max_iter=1000,C=1e12, random_state=0)
lr.fit(tf_x_train,Y_train)
y_pred=lr.predict(tf_x_test)

print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))
```

```python
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(tf_x_train,Y_train)
Y_predknn = classifier.predict(tf_x_test)

print("Accuracy:",metrics.accuracy_score(Y_test, Y_predknn))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
abc = AdaBoostClassifier(DecisionTreeClassifier(max_depth = 10), n_estimators=50,learning_rate=1)
model = abc.fit(tf_x_train,Y_train)
y_predabc = model.predict(tf_x_test)

print("Accuracy:",metrics.accuracy_score(Y_test, y_predabc))
```

```python
randomForest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)
modelRf = randomForest.fit(tf_x_train, Y_train)
Y_predRf = modelRf.predict(tf_x_test)

print("Accuracy:",metrics.accuracy_score(Y_test, Y_predRf))
```
